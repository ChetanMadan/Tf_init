```python
#LINEAR REGRESSION 
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt 

m = tf.Variable(0.)
b = tf.Variable(0.)

X=tf.placeholder(dtype=tf.float32) #at run time we will specify what constant value it will hold 
Y=tf.placeholder(dtype=tf.float32)

y_p= m*X +b
loss = tf.reduce_mean((y_p-Y)**2)

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.0001)
train = optimizer.minimize(loss)

init = tf.global_variables_initializer()

#create the dataset 

X_train = np.arange(50)
Y_train = 5*X_train + 8 + 10*(2*np.random.rand(50)-1)

plt.scatter(X_train, Y_train)
n_epochs = 1000

with tf.Session() as sess: 
     sess.run(init)
    
     for i  in range(n_epochs):
        train_e,loss_e,m_e,b_e= sess.run([train,loss,m,b],feed_dict={X: X_train,Y: Y_train})
        print('cost : {}'.format(loss_e))
    
        
    
# NON LINEAR REGRESSIONS 
# - add a new variable to the hypothesis function to cater to the quadratic- introduced virtual features - derived from known features  
        
```
