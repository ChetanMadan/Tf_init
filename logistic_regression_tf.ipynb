```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
mean1 = np.array([5,5])
cov1 = np.array([[1,0],[0,1]])
class1= np.random.multivariate_normal(mean1,cov1,500)
```

```python
mean2 = np.array([9,9])
cov2 = np.array([[1,0],[0,1]])
class2= np.random.multivariate_normal(mean2,cov2,500)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(class1[:,0],class1[:,1])
plt.scatter(class2[:,0],class1[:,1])
```

```python
data_X = np.append(class1,class2, axis=0)
data_Y =np.append(np.zeros(200),np.ones(200))
```

```python
import tensorflow as tf 
```

```python
X = tf.placeholder(dtype=np.float32)
Y = tf.placeholder(dtype=np.float32)

m = tf.Variable(tf.random_normal(shape=[2,1]))
b = tf.Variable(tf.random_normal(shape=[1,1]))

#hypothesis 
z = tf.matmul(X,m)+b
```

```python
#using sigmoid_cross_entropy_with_logits - logits 

loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=z, labels=Y))

optimiser = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimiser.minimize(loss)

init= tf.global_variables_initializer()
```

```python
#EXECUTION OF THE GRAPH 

n_epochs = 1000

with tf.Session() as sess:
    sess.run(init)
    
    for i in range(n_epochs):
        train_e, loss_e, m_e, b_e = sess.run([train,loss,m,b],feed_dict={X :data_X,Y :data_Y})
        print(loss_e)
    print("FINAL",loss_e,m_e,b_e)
                                             
```

```python
#plotting 

x = np.linspace(0,10.0,100)
y = np.linspace(0,10.0,100)
Xp,Yp = np.meshgrid(x,y)

F= Xp*m_e[0] + Yp*m_e[1] +b_e

plt.figure(figsize=(8,6))
plt.contour(Xp,Yp,F,0)

plt.scatter(class1[:,0],class1[:,1])
plt.scatter(class2[:,0],class2[:,1])

plt.show()
```
